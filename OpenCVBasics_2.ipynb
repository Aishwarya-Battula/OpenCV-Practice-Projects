```python
#drawing on an image
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img= cv.imread('cat.jpg')
cv2_imshow(img)
```

```python
#blank image to draw on
blank=np.zeros((500,500,3),dtype='uint8')
cv2_imshow(blank)
```

```python
#1. paint the image a certain color
blank[:]=0,255,0 #green image
cv2_imshow(blank)
```

```python
blank[200:300,400:450]=255,255,0 
cv2_imshow(blank)
```

```python
#2. Draw a rectangle
rec=cv.rectangle(blank,(0,0),(250,250),(0,0,255),thickness=2)
cv2_imshow(rec)
```

```python
rec=cv.rectangle(blank,(0,0),(250,250),(0,0,255),thickness=cv.FILLED)
cv2_imshow(rec)
```

```python
#3. Draw a circle

circle=cv.circle(blank,(250,250),100,(255,255,255),thickness=5 )
cv2_imshow(circle)
```

```python
circle=cv.circle(blank,(250,250),40,(255,255,255),thickness=5 )
cv2_imshow(circle)
```

```python
#3. Write text on an image

txt=cv.putText(blank,'Hello',(0,400),cv.FONT_HERSHEY_DUPLEX,1.0,(0,0,0),thickness=2)
cv2_imshow(txt)
```
