Common functions used in a OpenCV project

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
img=cv.imread('cat.jpg')
cv2_imshow(img)
```

```python
#Converting to grayscale
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
#Blurring an image
blur=cv.GaussianBlur(img,(7,7),cv.BORDER_DEFAULT)
cv2_imshow(blur)
```

```python
#Create an edge cascade- edge detection
canny=cv.Canny(img,125,175)
cv2_imshow(canny)
```

```python
#Dilating the image
dilated=cv.dilate(canny,(7,7),iterations=5)
cv2_imshow(dilated)
```

```python
#Eroding the image- getting back the sharp edges
eroded=cv.erode(dilated,(7,7),iterations=6)
cv2_imshow(eroded)
```

```python
#Resize
resized=cv.resize(img,(500,500))
cv2_imshow(resized)
```

```python
#Cropping
cropped=img[80:280, 150:330]
cv2_imshow(cropped)
```
